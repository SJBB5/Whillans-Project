```python
import pandas as pd
import numpy as np
```

```python
full_df = pd.read_csv("/Users/sambrown/Documents/SURF/10-18.csv")
inter_df = pd.read_csv("/Users/sambrown/Documents/SURF/inter_df.csv")
```

```python
full_df.head()
```

```python
full_df['start_time'] = pd.to_datetime(full_df['start_time'])
inter_df['time'] = pd.to_datetime(inter_df['time'])

full_df = full_df.sort_values('start_time')
inter_df = inter_df.sort_values('time')

# Merge 
merged_df = pd.merge_asof(
    full_df,
    inter_df[['time', 'disp_standardized', 'r2_standardized', 'slope_standardized']],
    left_on='start_time',
    right_on='time',
    direction='nearest',
    tolerance=pd.Timedelta('1H')  # Require times to be within 1 hour
)

merged_df = merged_df.drop(columns='time')
```
