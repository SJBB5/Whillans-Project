```python
# Sam Brown
# sam_brown@mines.edu
# June 17
# Goal: Use the station class to efficiently gather averages and other statistics for stations to allow for easier standardization

from Stations import Station
import my_lib.funcs
import pandas as pd
import numpy as np

# We will standardize base on all of the avaliable data and years we have
stations = [
    'la01', 'la02', 'la03', 'la04', 'la05', 'la06', 'la07', 'la08', 'la09', 'la10',
    'la11', 'la12', 'la13', 'la14', 'la15', 'la16', 'la17', 'la18',
    'gz01', 'gz02', 'gz03', 'gz04', 'gz05', 'gz06', 'gz07', 'gz08', 'gz09', 'gz10',
    'gz11', 'gz12', 'gz13', 'gz14', 'gz15', 'gz16', 'gz17', 'gz18', 'gz19', 'gz20',
    'mg01', 'mg02', 'mg03', 'mg04', 'mg05', 'mg06', 'mg07'
]

# Load in all years of data
df_2008 = ("/Users/sambrown04/Documents/SURF/Events/2008_2008Events2stas")
df_2009 = ("/Users/sambrown04/Documents/SURF/Events/2009_2009Events2stas")
df_2010 = ("/Users/sambrown04/Documents/SURF/Events/2010_2010Events2stas")
df_2011 = ("/Users/sambrown04/Documents/SURF/Events/2011_2011Events2stas")
df_2012 = ("/Users/sambrown04/Documents/SURF/Events/2012_2012Events2stas")
df_2013 = ("/Users/sambrown04/Documents/SURF/Events/2013_2013Events2stas")
df_2014 = ("/Users/sambrown04/Documents/SURF/Events/2014_2014Events2stas")
df_2015 = ("/Users/sambrown04/Documents/SURF/Events/2015_2015Events2stas")
df_2016 = ("/Users/sambrown04/Documents/SURF/Events/2016_2016Events2stas")
df_2017 = ("/Users/sambrown04/Documents/SURF/Events/2017_2017Events2stas")
df_2018 = ("/Users/sambrown04/Documents/SURF/Events/2018_2018Events2stas")
df_2019 = ("/Users/sambrown04/Documents/SURF/Events/2019_2019Events2stas")

years = [df_2008, df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019]
```

```python
station_objects = {}
for i, name in enumerate(stations, start = 1):
    print(f"[{i}/{len(stations)}] Creating station object for: {name}")
    station_objects[name] = Station(name, years)
```

```python
# Create DF that stores all of these values
stations_df = pd.DataFrame(columns = ['Station', 'pre-slip_area', 'pre-slip_area_sd', 'slip_severity', 'slip_severity_sd', 'slip_size', 'slip_size_sd', "x_cor", "y_cor"])

for station_name in station_objects:
    station = station_objects[station_name]
    stations_df.loc[len(stations_df)] = {
                "Station": station.name ,
                "pre-slip_area": station.pre_slip_a,
                "pre-slip_area_sd": station.pre_slip_a_sd,
                "slip_severity": station.slip_severity,
                "slip_severity_sd": station.slip_severity_sd,
                "slip_size": station.slip_size,
                "slip_size_sd": station.slip_size_sd,
                "x_cor": station.xcor,
                "y_cor": station.ycor
            }
```
